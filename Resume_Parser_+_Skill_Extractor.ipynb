```python
from google.colab import files
uploaded = files.upload()
```

```python
import PyPDF2
import docx

def extract_text(file_name):
    text = ""

    if file_name.endswith(".pdf"):
        with open(file_name, "rb") as f:
            reader = PyPDF2.PdfReader(f)
            for page in reader.pages:
                text += page.extract_text()

    elif file_name.endswith(".docx"):
        doc = docx.Document(file_name)
        for para in doc.paragraphs:
            text += para.text + "\n"

    return text

file_name = list(uploaded.keys())[0]
resume_text = extract_text(file_name)

print(resume_text[:1000])  # preview
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")

# Predefined skill list (you can expand)
skills_list = [
    "python", "java", "c++", "machine learning", "deep learning",
    "nlp", "sql", "excel", "power bi", "tableau",
    "django", "flask", "react", "angular", "aws", "azure"
]

def extract_skills(text):
    doc = nlp(text.lower())
    found_skills = []

    for token in doc:
        if token.text in skills_list:
            found_skills.append(token.text)

    return list(set(found_skills))

skills = extract_skills(resume_text)
print("Extracted Skills:", skills)
```

```python
import re

def extract_experience(text):
    pattern = r'(\d+)\+?\s*(years|yrs)'
    matches = re.findall(pattern, text.lower())

    if matches:
        return matches
    else:
        return "Not found"

experience = extract_experience(resume_text)
print("Experience:", experience)
```

```python
job_description = """
We are looking for a Python developer with experience in Django, SQL, and AWS.
"""

job_skills = extract_skills(job_description)

match = set(skills).intersection(set(job_skills))

print("Job Skills:", job_skills)
print("Matched Skills:", match)
print("Match %:", len(match)/len(job_skills)*100 if job_skills else 0)
```

```python
def suggest_improvements(skills):
    suggestions = []

    if "python" not in skills:
        suggestions.append("Consider learning Python")
    if "sql" not in skills:
        suggestions.append("Add SQL to your skillset")
    if "aws" not in skills:
        suggestions.append("Cloud skills (AWS/Azure) are recommended")

    return suggestions

print("Suggestions:", suggest_improvements(skills))
```

```python
print("\n===== FINAL RESULT =====")
print("Skills:", skills)
print("Experience:", experience)
print("Suggestions:", suggest_improvements(skills))
```
